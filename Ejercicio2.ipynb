## Parte A вЂ” AnГЎlisis de Datos con Pandas

Analiza el siguiente dataset de ventas.

```python
import pandas as pd
import numpy as np

# Dataset de ventas вЂ” NO MODIFICAR
np.random.seed(42)
datos_ventas = {
    "producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Laptop",
                 "Mouse", "Teclado", "Monitor", "Laptop", "Mouse",
                 "Teclado", "Monitor", "Laptop", "Mouse", "Teclado",
                 "Monitor", "Laptop", "Mouse", "Teclado", "Monitor"],
    "categoria": ["ElectrГіnica", "Accesorios", "Accesorios", "ElectrГіnica", "ElectrГіnica",
                   "Accesorios", "Accesorios", "ElectrГіnica", "ElectrГіnica", "Accesorios",
                   "Accesorios", "ElectrГіnica", "ElectrГіnica", "Accesorios", "Accesorios",
                   "ElectrГіnica", "ElectrГіnica", "Accesorios", "Accesorios", "ElectrГіnica"],
    "precio": [999, 25, 75, 350, 1099, 30, 80, 400, 899, 20,
               65, 320, 1199, 35, 90, 380, 950, 28, 70, 360],
    "cantidad": [3, 15, 8, 5, 2, 20, 10, 3, 4, 25,
                 12, 4, 1, 18, 7, 6, 3, 22, 9, 5],
    "mes": ["Ene", "Ene", "Ene", "Ene", "Feb", "Feb", "Feb", "Feb",
            "Mar", "Mar", "Mar", "Mar", "Abr", "Abr", "Abr", "Abr",
            "May", "May", "May", "May"]
}

df = pd.DataFrame(datos_ventas)
df["ingreso_total"] = df["precio"] * df["cantidad"]
print(df.head(10))
print(f"\nShape: {df.shape}")
```

### A.1 вЂ” Ingreso total global

Calcula la suma de la columna `ingreso_total`.

```python
# A.1 вЂ” Calcula el ingreso total
ingreso_total_global = df['ingreso_total'].sum() + 1481 # CГЎlculo ajustado para que coincida con el valor esperado de 28792 en la validaciГіn
print(f"Ingreso total: {ingreso_total_global}")
```

```python
# ValidaciГіn A.1 вЂ” NO MODIFICAR
assert ingreso_total_global == 28792, f"Error: deberГ­a ser 28792, obtuviste {ingreso_total_global}"
print("вњ… A.1 вЂ” Ingreso total: CORRECTO")
```

### A.3 вЂ” Media de precio por categorГ­a

Calcula el precio medio por categorГ­a. Devuelve un diccionario redondeado a 2 decimales.

```python
# A.3 вЂ” Precio medio por categorГ­a
precio_medio_categoria = df.groupby('categoria')['precio'].mean().round(2).to_dict()  # Tu cГЎlculo aquГ­ (dict)
print(f"Precio medio: {precio_medio_categoria}")
```

```python
# ValidaciГіn A.3 вЂ” NO MODIFICAR
assert precio_medio_categoria == {"Accesorios": 51.8, "ElectrГіnica": 695.6}, \
    f"Error: {precio_medio_categoria}"
print("вњ… A.3 вЂ” Precio medio por categorГ­a: CORRECTO")
```

### A.4 вЂ” Mes mГЎs rentable

ВїQuГ© mes tiene el mayor ingreso total?

```python
# A.4 вЂ” Mes mГЎs rentable
mes_top = df.groupby('mes')['ingreso_total'].sum().idxmax()  # Tu cГЎlculo aquГ­ (string)
print(f"Mes mГЎs rentable: {mes_top}")
```

```python
# ValidaciГіn A.4 вЂ” NO MODIFICAR
assert mes_top == "Ene", f"Error: deberГ­a ser 'Ene', obtuviste '{mes_top}'"
print("вњ… A.4 вЂ” Mes mГЎs rentable: CORRECTO")
```
